```python
import numpy as np
from layer import *
from model import *
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
x_train = np.load("x_train.npy")
y_train = np.load("y_train.npy")
x_test = np.load("x_test.npy")
y_test = np.load("y_test.npy")
```

```python
nn = model(x_train,y_train,96,alpha=0.01,accuracy=0.99)
layer1 = layer(2352,196)
layer3 = layer(196,96,activation='softmax')
nn.add(layer1)
nn.add(layer3)
nn.train()
nn.predict(x_test,y_test)
```

```python
fig = plt.figure()
plt.plot(nn.accuracy_history)
fig.suptitle('Accuracy vs Epochs', fontsize=20)
plt.xlabel('Number of epochs', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
```

```python
fig = plt.figure()
plt.plot(nn.loss_history)
fig.suptitle('Loss vs Epochs', fontsize=20)
plt.xlabel('Number of epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
```
